```python
import igraph as g
import gensim
```

```python
import pandas as pd 
df = pd.read_csv("~/Desktop/mobility_network_trajectories_real.csv")
df.head()

tokens = []
for sentence in df.x:
    tokens.append(sentence.split(' '))
tokens
```

```python
# Build and train the gensim word2vec model.
# First, perform initial training on unshuffled vocabulary
model = gensim.models.Word2Vec(
    tokens,
    size = 100,
    window = 1, # just use the entire sentence
    min_count = 5, # Remove tokens that don't appear enough
    workers = 10, # paralellize, use 4 workers
    ns_exponent = 1.0, # Gamma value for word2vec
    iter = 5,
    sg = 1 # use the skip_gram model
) # end model
```

```python
import numpy as np
records = []
for token in model.wv.vocab:
    records.append({
        "token": token,
        "norm": np.linalg.norm(model[token]),
        "size": model.wv.vocab[token].count
    })

df = pd.DataFrame(records)
df.to_csv("~/Desktop/mobility_network_trajectories_real_final.csv")
```
